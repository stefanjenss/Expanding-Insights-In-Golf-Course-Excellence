```python
"""
1. Load in the Dataset
"""
# Import the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "/Users/stefanjenss/Documents/Expanding-Insights-In-Golf-Course-Excellence/top_and_non_golf_course_reviews.xlsx"
df = pd.read_excel(file_path)

# Display the first 3 and last 3 rows of the dataset
df.head(3)
df.tail(3)
```

```python
"""
2. Explore Summary Statistics
"""
summary_stats = df.describe(include="all")

# Display the summary statistics
summary_stats
```

```python
"""
3. Explore distribution of the data in the dataset for variables of interest
"""
"""
3.1 Distribution of the courses by label (top-100 vs. non-top-100)
"""
label_distribution = df["label"].value_counts()

# Plot the distribution of the courses by label
plt.figure(figsize=(6, 4))
sns.set_style("darkgrid")
sns.set_palette("Set2")
sns.countplot(data=df, x="label", hue="label", edgecolor="black")
plt.title("Distribution of the Courses by Label")
plt.xlabel("Label")
plt.ylabel("Count")

# Export the plot as a high-resolution image
# plt.savefig("/Users/stefanjenss/Documents/Expanding-Insights-In-Golf-Course-Excellence/Figures/1_distribution_of_courses_by_label.png", dpi=300)

plt.show()
```

```python
"""
3.2 Distribution of the year built for the courses, by label
"""
# Plot the distribution of the year built for the courses, by label
plt.figure(figsize=(9, 6))
sns.set_style("darkgrid")
sns.set_palette("Set2")
sns.histplot(data=df, x="year_built", hue="label", edgecolor="black", multiple="stack")
plt.title("Distribution of the Year Built for the Courses, by Label")
plt.xlabel("Year Built")
plt.ylabel("Count")

# Export the plot as a high-resolution image
# plt.savefig("/Users/stefanjenss/Documents/Expanding-Insights-In-Golf-Course-Excellence/Figures/2_distribution_of_year_built_by_label.png", dpi=300)

plt.show()
```

```python
"""
3.3 Distribution of the review authors by label
"""
# Plot the distribution of the review authors by label
plt.figure(figsize=(9, 8))
sns.set_style("darkgrid")
sns.set_palette("Set2")
sns.histplot(data=df, y="review_author", hue="label", edgecolor="black", multiple="stack")
plt.title("Distribution of the Review Authors by Label")
plt.xlabel("Count")
plt.ylabel("Review Author")
# Adjust layout to prevent cutoff
plt.tight_layout()
# Export the plot as a high-resolution image
plt.savefig("/Users/stefanjenss/Documents/Expanding-Insights-In-Golf-Course-Excellence/Figures/3_distribution_of_review_authors_by_label.png", dpi=300)

plt.show()
```

```python
"""
3.4 Distribution of golf course architect by label
"""
# Plot the distribution of the golf course architect by label
plt.figure(figsize=(9, 8))
sns.set_style("darkgrid")
sns.set_palette("Set2")
sns.histplot(data=df, y="architect", hue="label", edgecolor="black", multiple="stack")
plt.title("Distribution of the Golf Course Architect by Label")
plt.xlabel("Count")
plt.ylabel("Golf Course Architect")
# Adjust layout to prevent cutoff
plt.tight_layout()
# Export the plot as a high-resolution image
plt.savefig("/Users/stefanjenss/Documents/Expanding-Insights-In-Golf-Course-Excellence/Figures/4_distribution_of_golf_course_architect_by_label.png", dpi=500)

plt.show()
```
